```python
# 生成jsonl文件
import os
import json
modality = 'MRI'
# organ = ["liver", "spleen", "kidney"]
organ_MRI = ["liver"]
if os.path.exists('abdomen_demo_MRI.jsonl'):
    os.remove('abdomen_demo_MRI.jsonl')
    
path = './data_tosegmented/data_mri/'
IDs = ['01190626V001', '01190718V003', '01200408V005', '01200514V002']
# MRI
for ID in sorted(IDs):
    # if 已经有了分割结果
        dic = {"image": path + '/'+ID+'.nii.gz', "label": organ_MRI, "modality": modality, "dataset":"abdomen"}
        json_data = json.dumps(dic, ensure_ascii=False)

        # 打开文件
        file = open('abdomen_demo_MRI.txt', 'a')

        # 写入 JSON 字符串
        file.write(json_data)
        file.write('\n')

                # 关闭文件
        file.close()
os.system('mv abdomen_demo_MRI.txt abdomen_demo_MRI.jsonl')

```
