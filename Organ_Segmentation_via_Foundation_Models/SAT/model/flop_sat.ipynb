```python
from thop import profile

from maskformer import Maskformer


model = Maskformer().cuda()    
image = torch.rand((1, 3, 512, 512, 1)).cuda()
query = torch.rand((2, 10, 768)).cuda()
segmentations = model(query, image)
print(segmentations.shape)
    
# model = Maskformer(args.vision_backbone, args.crop_size, args.patch_size, args.deep_supervision)
model = Maskformer(vision_backbone='UNET', image_size=[512, 512, 1], patch_size=[32, 32, 32], deep_supervision=False)
model = model.to("cuda:0")


shape = (1, 3, 512, 512)  # 定义形状，例如 3行4列
tensor = torch.rand(shape)  # 均匀分布
flops, params = profile(model, inputs=(image, query))
print(flops/1e9, 'G',params/1e6,'M')
return
```
